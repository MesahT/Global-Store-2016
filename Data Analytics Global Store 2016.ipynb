# **1. DATA CLEANING**

```python
#Importing the necessary libraries
import numpy as np 
import pandas as pd 
import datetime as datetime 
```

```python
#importing our data and storing into the variable global_store 
global_store = pd.read_csv('global_superstore_2016.csv')
```

```python
#displaying 5 rows of our data 
global_store.head(5)
```

**Cleaning the data**

```python
#Finding information about our data, ie columns datatypes 
global_store.columns
```

```python
#Converting Order Date to datetime format 
order_date = global_store['Order Date'].tolist()
order_datetime = pd.to_datetime(order_date)
order_datetime
```

```python
#Converting the data to a series so that we can concatenate it to the global store data
Order_Date = pd.Series(order_datetime, name = 'Order Datetime Series')
Order_Date
```

```python
#Coverting the ship date to datetime format 
ship_date = global_store['Ship Date'].tolist()
ship_datetime = pd.to_datetime(ship_date)
Ship_Date = pd.Series(ship_datetime, name = 'Ship Datetime series')
Ship_Date
```

```python
#Converting Sales from string data format to float 
sales_list = global_store['Sales'].tolist()
sales = pd.Series(sales_list, name = 'Sales Series')
sales
```

```python
#Converting sales series to a numeric (float) datatype 
Sales = pd.to_numeric(sales.replace('[$,]', '', regex = True))
Sales
```

```python
#Converting Profit data from string to float 
profit_list = global_store['Profit'].tolist()
profit_series = pd.Series(profit_list, name = 'Profit Series')
profit_series
```

```python
#Converting the profit series to a numeric (float) data type
Profit = pd.to_numeric(profit_series.replace('[$,()]', '', regex = True))
Profit
```

**Compiling our newly cleaned dataset to work with**

```python
#Concatenating the important columns to work with 
store = pd.concat([global_store[['Row ID', 'Order ID']], Order_Date, Ship_Date, global_store[['Ship Mode',
       'Customer ID', 'Customer Name', 'Segment', 'Postal Code', 'City',
       'State', 'Country', 'Region', 'Market', 'Product ID', 'Category',
       'Sub-Category', 'Product Name']], Sales, Profit, global_store[['Quantity', 'Discount', 'Shipping Cost', 'Order Priority']]
], axis = 1)
store.head(2)
```

# **Overall Ranking of India where Profit is maximum under the Category 'Technology'**

```python
#Grouping the Profit Series and Category Technology by Country 
condition_rank = (store['Category'] == 'Technology')
grouped = store[['Country', 'Profit Series']][condition_rank].groupby(by = 'Country').sum()
grouped
```

```python
#Using the ranking function to rank by maximum profit with the highest as rank number 1, thus in descending order
grouped['Profit Rank'] = grouped['Profit Series'].rank(method = 'max', ascending = False)
grouped
```

```python
#Now sorting the ranked data by Profit Rank in ascending order
ranked = grouped.sort_values(by = 'Profit Rank')
ranked
```

```python
#Searching for the Overall rank of India in terms of maximum profits in the ranked data 
ranked[ranked.index == 'India']
```
